```python
import numpy as np
```

```python
def arrangements(n: int, k: int) -> int:
    """Число размещений.
    """
    
    return np.math.factorial(n) // np.math.factorial(n - k)
```

```python
def permutations(n: int) -> int:
    """Число перестановок.
    """
    
    return np.math.factorial(n)
```

```python
def combinations(n: int, k: int) -> int:
    """Число сочетаний.
    """
    
    return np.math.factorial(n) // (np.math.factorial(k) * np.math.factorial(n - k))
```

#### Задача 2

Семь человек рассаживаются наудачу на скамейке. Какова вероятность того, что рядом будут сидеть: а) два определённых человека? б) три определённых человека?

Пробуем создать функцию, которая будет рассчитывать вероятность в зависимости от количества человек в зависимой группе.
Искомая вероятность это отношение  количество успешных размещений к количеству возможных размещений.
Нам следует определить количество мест на скамейке исходя из того, что группа занимает одно место 
(количество независимых мест). 
Далее рассчитываем количество перестановок в независимых местах, количество перестановок внутри группы.
Также рассчитываем возможные перестановки на местах, не занятых определенными людьми
Таким образом количество успешных размещений группы = количество размещений в общем количестве человек*размещения внутри группы*размещения независимых мест
Количество возможных размещений - это кол-во перестановок в общем кол-ве людей на скамейке.

Может как то запутанно вышло, но хотелось сделать функцию "на все случаи жизни"

```python
def bench (n: int, k: int) -> int:   #n - общее количество человек, k - количество определенных человек
    
    dep_pos = k - 1    #количество зависимых мест 
       
    indep_pos = n - dep_pos    #количество независимых мест 

    c_indep = arrangements(indep_pos,1)    #число размещений независимых мест:
    
    gr_perm = permutations(k)    # число возможных перестановок в группе определенных человек

    free_pos = n - k    # свободные места для неопределенных людей
    
    free_perm = permutations(free_pos)    #число возможных перестановок в свободных местах
    
    tot_perm = permutations (n)    #возможное число перестановок общего количества человек
    
    c_succ = c_indep * gr_perm * free_perm    # количество успешных размещений        
        
    prob = c_succ/tot_perm    # искомая вероятность
    
    return prob
```

```python
print("вариант а: ", bench(7,2), "\nвариант б: ", bench(7,3))
```

#### Задача 3

Из 60 вопросов, входящих в экзаменационные билеты, студент знает 50. Какова вероятность того, что среди трёх наугад выбранных вопросов студент знает: а) все? б) два?

```python
"""
tot_quest = 60 всего вопросов 
know_quest = 50 известные вопросы
rand_quest = 3 количество вопросов наугад
know_ans = [3,2] количество известных ответов

Искомая вероятность это отношение 
    произведения
    (комбинации выбора 2 или 3 известных вопросов из множества известных вопросов) 
    и 
    (комбинации выбора 1 или 0 неизвестных вопросов из неизвестных вопросов)
к 
комбинации выбора 3-х вопросов из всех возможых вопросов

"""

def exam_succ (tot_quest: int, know_quest: int, rand_quest: int, know_ans) -> int:

    comb_tot_var = combinations(tot_quest, rand_quest) # общее кол-во комбинаций выбора вопросов в билете
    
    comb_know_ans = combinations(know_quest, know_ans)   # кол-во комбинаций выбора известного ответа
    
    unknow_quest = tot_quest - know_quest   # кол-во вопросов, которые студент не знает
    unknow_ans = rand_quest - know_ans   # кол-во неверных ответов в наугад выбранных
    comb_unknow_ans = combinations(unknow_quest, unknow_ans)   # кол-во комбинаций ответов на неизвестные вопросы
   
    prob_answ = comb_know_ans * comb_unknow_ans / comb_tot_var   # искомая вероятность 
    return prob_answ
```

```python
print("вариант а: ", exam_succ(60, 50, 3, 3), "\nвариант б :", exam_succ(60, 50, 3, 2))
```

#### Задача 5

Допустим, имеется некоторая очень редкая болезнь (поражает 0.1% населения). Вы приходите к врачу, вам делают тест на эту болезнь, и тест оказывается положительным. Врач говорит вам, что этот тест верно выявляет 99% больных этой болезнью и всего лишь в 1% случаев даёт ложный положительный ответ.

Вопрос: какова вероятность, что вы действительно больны ей?

Подсказка: вновь используйте формулу Байеса с раскрытием знаменателя с помощью формулы полной вероятности.

А - болен

B - тест положителен

P(A|B) - вероятность что болен, при условии, что тест положителен

P(B|A) - вероятность что тест положителен, при условии, что болен = 0,99

P(A) - вероятность того, что болен = 0,001

P(B) - вероятность положительного теста (=вер-ть положит теста у больного + вероятность ложного теста у здорового)было сложнее всего додуматься

P(B)= P(B|A)*P(A)+P(B|notA)*P(notA)

P(A|B)=P(B|A)*P(A)/P(B)

P(A|B)=P(B|A)*P(A)/(P(B|A)*P(A)+P(B|notA)*P(notA))

```python
prob_sick_pos = 0.001

prob_sick_neg = 0.999

prob_test_wrong = 0.01

prob_test_right = 0.99

prob = prob_test_right*prob_sick_pos/(prob_test_right*prob_sick_pos+prob_test_wrong*prob_sick_neg)

print(prob)
```
